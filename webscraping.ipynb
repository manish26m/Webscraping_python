```python
import os

def create_folder_structure(base_path):
    folders = [
        'Schools', 'Clubs', 'Maps', 'Officials/BDO', 'Officials/IDO', 
        'Post_Offices', 'Hospitals', 'Demographics'
    ]
    for folder in folders:
        os.makedirs(os.path.join(base_path, folder), exist_ok=True)

create_folder_structure('Block_Rosopunj_Info')
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os

# URL of the webpage that lists government schools in Raspunja
url = "https://schools.org.in/west-bengal/south--twenty-four-pargan/thakurpukur-maheshtala/rasapunja-pk-high-school"

# Send a GET request to the webpage
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the webpage content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the table with class 'table table-striped' and extract school names
    table = soup.find('table', {'class': 'table table-striped'})
    school_names = [a.text for a in table.find_all('a')]

    # Create a DataFrame from the extracted school names
    df = pd.DataFrame(school_names, columns=['School Name'])

    # Create the directory if it doesn't exist
    output_dir = 'Block_Rosopunj_Info/Schools'
    os.makedirs(output_dir, exist_ok=True)

    # Save the DataFrame to a CSV file
    output_file = os.path.join(output_dir, 'schools.csv')
    df.to_csv(output_file, index=False)

    print(f"School names saved to {output_file}")
else:
    print("Failed to retrieve the webpage")
```

# POST OFFICE WORK:-


```python
import requests
import pandas as pd

def fetch_post_offices(pincode, output_file):
    url = f'https://api.postalpincode.in/pincode/{pincode}'
    response = requests.get(url)
    post_offices = response.json()

    data = []
    for office in post_offices[0]['PostOffice']:
        data.append({
            'Name': office['Name'],
            'BranchType': office['BranchType'],
            'DeliveryStatus': office['DeliveryStatus'],
            'District': office['District'],
            'Division': office['Division'],
        })

    df = pd.DataFrame(data)
    df.to_csv(output_file, index=False)

pincode = '700104'  # Example pincode
output_file = 'Block_Rosopunj_Info/Post_Offices/post_offices.csv'
fetch_post_offices(pincode, output_file)
```
